```python
from torchvision import models
from torch import nn
from torchsummary import summary
```

```python
model_resnet = models.resnet50(pretrained=False)
model_mobilenet = models.mobilenet_v2(pretrained=False)
model_resnext = models.resnext50_32x4d(pretrained=False)
model_inception = models.inception_v3(pretrained=False)
model_vgg16 = models.vgg16(pretrained=False)
```

```python
model_resnet.cuda()
summary(model_resnet, (3, 224, 224))
```

```python
model_mobilenet.cuda()
summary(model_mobilenet, (3, 224, 224))
```

```python
model_resnext.cuda()
summary(model_resnext, (3, 224, 224))
```

```python
model_vgg16.cuda()
summary(model_vgg16, (3, 224, 224))
```

```python
model_resnet.conv1 = nn.Conv2d(4, 64, kernel_size=7, stride=2, padding=3, bias=False)
nn.init.kaiming_normal_(model_resnet.conv1.weight, mode='fan_out', nonlinearity='relu')
model_resnet.fc = nn.Linear(model_resnet.fc.in_features, 1200)
```

```python
model_mobilenet.features[0][0] = nn.Conv2d(4, 32, kernel_size=3, stride=2, padding=1, bias=False)
nn.init.kaiming_normal_(model_mobilenet.features[0][0].weight, mode='fan_out', nonlinearity='relu')
model_mobilenet.classifier[1] = nn.Linear(model_mobilenet.classifier[1].in_features, 1200)
```

```python
model_vgg16.features[0] = nn.Conv2d(4, 64, kernel_size=3, stride=1, padding=1, bias=False)
nn.init.kaiming_normal_(model_vgg16.features[0].weight, mode='fan_out', nonlinearity='relu')
model_vgg16.classifier[6] = nn.Linear(model_vgg16.classifier[6].in_features, 1200)
```

```python
model_resnext.conv1 = nn.Conv2d(4, 64, kernel_size=7, stride=2, padding=3, bias=False)
nn.init.kaiming_normal_(model_resnext.conv1.weight, mode='fan_out', nonlinearity='relu')
model_resnext.fc = nn.Linear(model_resnext.fc.in_features, 1200)
```

```python
model_resnet.cuda()
summary(model_resnet, (4, 224, 224))
```
